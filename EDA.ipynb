# 📊 Air Quality EDA
Exploratory Data Analysis on air quality and weather dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
data = pd.read_csv("../data/historical_aqi.csv")
data.head()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between Features")
plt.show()
```

```python
# AQI distribution
plt.hist(data['AQI'], bins=30, color='skyblue', edgecolor='black')
plt.title("AQI Distribution")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()
```

```python
# Time series of AQI
data['Date'] = pd.to_datetime(data['Date'])
plt.figure(figsize=(12,6))
plt.plot(data['Date'], data['AQI'], color='green')
plt.title("AQI Over Time")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.grid(True)
plt.show()
```
